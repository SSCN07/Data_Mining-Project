```python
import requests
from bs4 import BeautifulSoup
import time

baseurl = 'https://ci.coinafrique.com/categorie/voitures?page='

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

car_data = []
max_items = 1000
current_page = 1

print("‚è≥ Scraping cars...")

while len(car_data) < max_items:
    url = baseurl + str(current_page)
    print(f"\nüìÑ Page {current_page}: {url}")
    
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.content, 'lxml')
    
    productlist = soup.find_all('div', class_='col s6 m4 l3')
    
    if not productlist:
        print("‚ö†Ô∏è No more products found.")
        break
    
    for card in productlist:
        # Car name
        name_tag = card.find('p', class_='ad__card-description')
        name = name_tag.text.strip() if name_tag else "N/A"
        
        # Price
        price_tag = card.find('p', class_='ad__card-price')
        price = price_tag.text.strip().replace('\n', ' ') if price_tag else "N/A"
        
        # Seller name
        seller_tag = card.find('p', class_='title')
        seller = seller_tag.text.strip() if seller_tag else "N/A"
        
        # Location
        location_tag = card.find('p', class_='ad__card-location')
        location = location_tag.text.strip() if location_tag else "N/A"
        
        # Date posted
        date_tag = card.find('div', class_='card-title ad__card-timesince')
        if date_tag:
            spans = date_tag.find_all('span')
            if len(spans) >= 2:
                date_posted = spans[0].text.strip() + ' ' + spans[1].text.strip()
            else:
                date_posted = "N/A"
        else:
            date_posted = "N/A"
        
        # --- NEW: Extract car image ---
        # Method 1: Look for img tag (most common)
        img_tag = card.find('img')
        if img_tag:
            # Try to get 'src' attribute first, if not found try 'data-src' (lazy loading)
            image_url = img_tag.get('src') or img_tag.get('data-src') or "N/A"
        else:
            image_url = "N/A"
        
        # Add the extracted information including image URL
        car_data.append({
            'name': name,
            'price': price,
            'seller': seller,
            'location': location,
            'date_posted': date_posted,
            'image_url': image_url  # NEW: Image URL added
        })
        
        if len(car_data) >= max_items:
            break
    
    current_page += 1
    time.sleep(1)

# üéâ Done - Print results
print(f"\n‚úÖ Scraped {len(car_data)} cars:\n")
for i, car in enumerate(car_data, 1):
    print(f"{i}. {car['name']}")
    print(f"   Price      : {car['price']}")
    print(f"   Seller     : {car['seller']}")
    print(f"   Location   : {car['location']}")
    print(f"   Date Posted: {car['date_posted']}")
    print(f"   Image URL  : {car['image_url']}\n")  # NEW: Display image URL
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Create transaction simulation function
def simulate_transactions(df, num_transactions=500):
    transactions = []
    
    for i in range(num_transactions):
        # Randomly select a car
        car = df.sample(1).iloc[0]#locte a row randomly
        
        # Generate random transaction details
        transaction_id = f"TXN{random.randint(10000, 99999)}"
        
        # Random buyer name
        buyers = ["Jean Koffi", "Marie Bamba", "Paul Kouassi", "Amina Traore", 
                 "Pierre Gnahore", "Fatou Diarra", "Marc Yao", "Sophie Kon√©",
                 "David Toure", "Sarah Camara"]
        buyer = random.choice(buyers)
        
        # Random transaction date (within last 30 days)
        days_ago = random.randint(0, 30)
        transaction_date = datetime.now() - timedelta(days=days_ago)
        
        # Random transaction amount (slightly negotiated price)
        price_str = car['price'].replace('CFA', '').replace(' ', '')
        try:
            base_price = float(price_str)
            # Add some negotiation variance (-10% to +5%)
            negotiation_factor = random.uniform(0.9, 1.05)
            transaction_amount = int(base_price * negotiation_factor)
        except:
            transaction_amount = random.randint(1000000, 50000000)
        
        # Payment methods
        payment_methods = ["Cash", "Bank Transfer", "Mobile Money", "Credit Card"]
        payment_method = random.choice(payment_methods)
        
        # Transaction status
        status = random.choice(["Completed", "Pending", "Processing"])
        
        transactions.append({
            'transaction_id': transaction_id,
            'car_brand': car['brand'],
            'original_price': car['price'],
            'transaction_amount': f"{transaction_amount:,}CFA",
            'buyer': buyer,
            'seller': car['seller'],
            'transaction_date': transaction_date.strftime('%Y-%m-%d'),
            'payment_method': payment_method,
            'status': status,
            'location': car['location'].replace('location_on', ''),
            'car_image_url': car['image_url']
        })
    
    return pd.DataFrame(transactions)

# Generate 500 transactions
transactions_df = simulate_transactions(df_car, 500)

# Display first 10 transactions
print("First 10 Simulated Transactions:")
print("=" * 80)
transactions_df.head(10)
```

```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import random

def create_market_basket_data(transactions_df):
    market_baskets = []
    
    for _, transaction in transactions_df.iterrows():
        basket = []
        
        # 1. Car Brand (with error handling)
        try:
            brand = str(transaction['car_brand']).split()[0]
            basket.append(f"Brand_{brand}")
        except:
            basket.append("Brand_Unknown")
        
        # 2. Price Category (with error handling)
        try:
            price_str = str(transaction['original_price']).replace('CFA', '').replace(' ', '').replace(',', '')
            price_val = float(price_str)
            if price_val < 5000000: 
                basket.append("Price_Budget")
            elif price_val < 15000000: 
                basket.append("Price_MidRange")
            elif price_val < 30000000: 
                basket.append("Price_Premium")
            else: 
                basket.append("Price_Luxury")
        except:
            basket.append("Price_Unknown")
        
        # 3. Payment Method (with error handling)
        try:
            payment = str(transaction['payment_method'])
            basket.append(f"Payment_{payment}")
        except:
            basket.append("Payment_Unknown")
        
        # 4. Location (with error handling)
        try:
            location = str(transaction['location']).split(',')[0]
            basket.append(f"Location_{location}")
        except:
            basket.append("Location_Unknown")
        
        # 5. Transaction Status (with error handling)
        try:
            status = str(transaction['status'])
            basket.append(f"Status_{status}")
        except:
            basket.append("Status_Unknown")
        
        # 6. Seller Type (FIXED - with proper type checking)
        try:
            seller = str(transaction['seller']).lower()  # Convert to string first
            if 'dealer' in seller or 'auto' in seller or 'car' in seller:
                seller_type = "Dealer"
            elif 'company' in seller or 'services' in seller or 'ltd' in seller:
                seller_type = "Company"
            else:
                seller_type = "Individual"
            basket.append(f"Seller_{seller_type}")
        except:
            basket.append("Seller_Individual")  # Default fallback
        
        # 7. Add random services (always works)
        services = ['Insurance', 'Warranty', 'Financing', 'Maintenance', 'Delivery']
        selected_services = random.sample(services, random.randint(1, 3))
        basket.extend([f"Service_{s}" for s in selected_services])
        
        # 8. Car Type/Category
        try:
            car_brand = str(transaction['car_brand']).lower()
            if 'range rover' in car_brand or 'bmw x' in car_brand or 'mercedes' in car_brand:
                car_type = "Luxury"
            elif 'rav4' in car_brand or 'tucson' in car_brand or 'sorento' in car_brand:
                car_type = "SUV"
            elif 'civic' in car_brand or 'corolla' in car_brand or 'elantra' in car_brand:
                car_type = "Sedan"
            elif 'golf' in car_brand or 'polo' in car_brand or 'swift' in car_brand:
                car_type = "Compact"
            else:
                car_type = "Other"
            basket.append(f"Type_{car_type}")
        except:
            basket.append("Type_Other")
        
        market_baskets.append(basket)
    
    return market_baskets

# 2. Transform your data (NOW IT SHOULD WORK)
market_baskets = create_market_basket_data(transactions_df)

# Display sample baskets
print("Sample Market Basket Transactions:")
print("=" * 60)
for i, basket in enumerate(market_baskets[:5]):
    print(f"Basket {i+1}: {basket}")
```

```python
# Create one-hot encoded DataFrame
def create_one_hot_encoded_data(transactions):
    # Initialize TransactionEncoder
    te = TransactionEncoder()
    
    # Fit and transform the transactions
    te_ary = te.fit(transactions).transform(transactions)
    
    # Create DataFrame
    one_hot_df = pd.DataFrame(te_ary, columns=te.columns_)
    
    return one_hot_df

# Generate one-hot encoded data
one_hot_df = create_one_hot_encoded_data(market_baskets)

print("One-Hot Encoded Data Shape:", one_hot_df.shape)
print("\nFirst 10 transactions (one-hot encoded):")
print(one_hot_df.head(10))

print("\nItem frequencies:")
item_frequencies = one_hot_df.sum().sort_values(ascending=False)
print(item_frequencies.head(20))
```

```python
# Alternative 1: Simple feature set
def create_simple_one_hot(df, num_transactions=500):
    simple_transactions = []
    
    for i in range(num_transactions):
        car = df.sample(1).iloc[0]
        car_brand = car['brand']
        
        # Basic features
        features = []
        
        # Price category
        try:
            price_val = float(car['price'].replace('CFA', '').replace(' ', ''))
            if price_val < 5000000:
                features.append('Low_Price')
            elif price_val < 15000000:
                features.append('Medium_Price')
            else:
                features.append('High_Price')
        except:
            features.append('Medium_Price')
        
        # Car type
        if any(x in car_brand.lower() for x in ['rav4', 'tucson', 'sportage', 'suv']):
            features.append('SUV')
        elif any(x in car_brand.lower() for x in ['corolla', 'civic', 'elantra', 'sedan']):
            features.append('Sedan')
        elif any(x in car_brand.lower() for x in ['hilux', 'ranger', 'pickup']):
            features.append('Truck')
        else:
            features.append('Other_Type')
        
        # Brand
        main_brand = car_brand.split(' ')[0]
        features.append(f'Brand_{main_brand}')
        
        # Location
        location = car['location'].replace('location_on', '').split(',')[0]
        features.append(f'Loc_{location}')
        
        # Seller type
        seller = car['seller']
        if seller in ['Kouame Luc Yao', 'Mr marco feh', 'Diomande']:
            features.append('Seller_Individual')
        elif seller in ['First Parc Auto', 'Auto Boss Compagnie', 'JS Voiture']:
            features.append('Seller_Dealer')
        else:
            features.append('Seller_Company')
        
        simple_transactions.append(features)
    
    # Create one-hot encoding
    te = TransactionEncoder()
    te_ary = te.fit(simple_transactions).transform(simple_transactions)
    simple_one_hot = pd.DataFrame(te_ary, columns=te.columns_)
    
    return simple_one_hot

# Alternative 2: Detailed feature set
def create_detailed_one_hot(df, num_transactions=500):
    detailed_transactions = []
    
    for i in range(num_transactions):
        car = df.sample(1).iloc[0]
        car_brand = car['brand']
        
        features = []
        
        # Price segments
        try:
            price_val = float(car['price'].replace('CFA', '').replace(' ', ''))
            if price_val < 3000000:
                features.extend(['Segment_Budget', 'Price_Under_3M'])
            elif price_val < 8000000:
                features.extend(['Segment_Economy', 'Price_3M_8M'])
            elif price_val < 15000000:
                features.extend(['Segment_MidRange', 'Price_8M_15M'])
            elif price_val < 30000000:
                features.extend(['Segment_Premium', 'Price_15M_30M'])
            else:
                features.extend(['Segment_Luxury', 'Price_Over_30M'])
        except:
            features.extend(['Segment_MidRange', 'Price_8M_15M'])
        
        # Vehicle type
        brand_lower = car_brand.lower()
        if any(x in brand_lower for x in ['rav4', 'tucson', 'sportage', 'santa fe', 'cr-v', 'suv']):
            features.append('Type_SUV')
        elif any(x in brand_lower for x in ['corolla', 'civic', 'elantra', 'camry', 'accord', 'sedan']):
            features.append('Type_Sedan')
        elif any(x in brand_lower for x in ['hilux', 'ranger', 'navara', 'l200', 'pickup']):
            features.append('Type_Truck')
        elif any(x in brand_lower for x in ['yaris', 'i10', 'swift', 'rio', 'compact']):
            features.append('Type_Compact')
        elif any(x in brand_lower for x in ['mercedes', 'bmw', 'audi', 'lexus', 'range']):
            features.append('Type_Luxury')
        else:
            features.append('Type_Other')
        
        # Brand origin
        japanese_brands = ['Toyota', 'Honda', 'Nissan', 'Mitsubishi', 'Suzuki', 'Mazda']
        korean_brands = ['Hyundai', 'Kia']
        european_brands = ['Mercedes', 'BMW', 'Audi', 'Volkswagen', 'Renault', 'Peugeot']
        
        main_brand = car_brand.split(' ')[0]
        if main_brand in japanese_brands:
            features.extend(['Origin_Japanese', f'Brand_{main_brand}'])
        elif main_brand in korean_brands:
            features.extend(['Origin_Korean', f'Brand_{main_brand}'])
        elif main_brand in european_brands:
            features.extend(['Origin_European', f'Brand_{main_brand}'])
        else:
            features.extend(['Origin_Other', f'Brand_{main_brand}'])
        
        # Age category
        try:
            car_year = int(''.join(filter(str.isdigit, car_brand))[-4:])
            current_year = datetime.now().year
            age = current_year - car_year
            if age <= 2:
                features.append('Age_New')
            elif age <= 5:
                features.append('Age_Young')
            elif age <= 10:
                features.append('Age_Mid')
            else:
                features.append('Age_Old')
        except:
            features.append('Age_Mid')
        
        # Transmission (inferred)
        if 'automatic' in brand_lower or 'auto' in brand_lower:
            features.append('Transmission_Auto')
        else:
            features.append('Transmission_Manual')
        
        # Drive type
        if any(x in brand_lower for x in ['4x4', 'awd', '4wd']):
            features.append('Drive_4WD')
        else:
            features.append('Drive_2WD')
        
        detailed_transactions.append(features)
    
    # Create one-hot encoding
    te = TransactionEncoder()
    te_ary = te.fit(detailed_transactions).transform(detailed_transactions)
    detailed_one_hot = pd.DataFrame(te_ary, columns=te.columns_)
    
    return detailed_one_hot

# Generate all versions
print("Generating different one-hot encoded datasets...")

# Simple version
simple_one_hot = create_simple_one_hot(df_car, 500)
print(f"Simple one-hot shape: {simple_one_hot.shape}")

# Detailed version
detailed_one_hot = create_detailed_one_hot(df_car, 500)
print(f"Detailed one-hot shape: {detailed_one_hot.shape}")

# Display sample of each
print("\nSimple one-hot sample (first 5 transactions, first 10 items):")
print(simple_one_hot.iloc[:5, :10])

print("\nDetailed one-hot sample (first 5 transactions, first 15 items):")
print(detailed_one_hot.iloc[:5, :15])
```

```python
# Apply Apriori algorithm to find frequent itemsets
def find_frequent_itemsets(one_hot_data, min_support=0.1, use_colnames=True):
    """
    Find frequent itemsets using Apriori algorithm
    """
    frequent_itemsets = apriori(one_hot_data, 
                               min_support=min_support, 
                               use_colnames=use_colnames)
    
    # Add length of itemsets
    frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
    
    return frequent_itemsets.sort_values('support', ascending=False)

# Find frequent itemsets for each dataset
print("Finding frequent itemsets...")

# For simple dataset
frequent_simple = find_frequent_itemsets(simple_one_hot, min_support=0.15)
print(f"Simple - Found {len(frequent_simple)} frequent itemsets")
print("Top 10 frequent itemsets (Simple):")
print(frequent_simple.head(10))

# For detailed dataset
frequent_detailed = find_frequent_itemsets(detailed_one_hot, min_support=0.1)
print(f"\nDetailed - Found {len(frequent_detailed)} frequent itemsets")
print("Top 10 frequent itemsets (Detailed):")
print(frequent_detailed.head(10))

# For market basket dataset
frequent_market = find_frequent_itemsets(one_hot_df, min_support=0.1)
print(f"\nMarket Basket - Found {len(frequent_market)} frequent itemsets")
print("Top 10 frequent itemsets (Market Basket):")
print(frequent_market.head(10))
```

```python
import requests
from bs4 import BeautifulSoup
import time
import csv

# TODO: Replace with actual Jumia URL
baseurl = 'https://www.jumia.ci/peripheriques-logiciels-accessoires/?q=telephone#catalog-listing'  # Adjust this URL

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'fr-FR,fr;q=0.9,en-US;q=0.8,en;q=0.7',
}

product_data = []
max_items = 1000
current_page = 1

print("‚è≥ Scraping Jumia computer accessories...")

while len(product_data) < max_items:
    # Jumia typically uses ?page=1 format
    if current_page == 1:
        url = baseurl
    else:
        url = f"{baseurl}?page={current_page}"
    
    print(f"\nüìÑ Page {current_page}: {url}")
    
    try:
        r = requests.get(url, headers=headers, timeout=15)
        r.raise_for_status()
        soup = BeautifulSoup(r.content, 'lxml')
        
        # Jumia product structure (may need adjustment based on actual HTML)
        productlist = soup.find_all('article', class_='prd')
        
        # Alternative selectors if above doesn't work
        if not productlist:
            productlist = soup.find_all('div', {'data-id': True})
        
        if not productlist:
            print("‚ö†Ô∏è No more products found.")
            break
        
        print(f"   Found {len(productlist)} products on this page")
        
        for product in productlist:
            # Product name
            name_tag = product.find('h3', class_='name') or product.find('div', class_='name')
            name = name_tag.text.strip() if name_tag else "N/A"
            
            # Price
            price_tag = product.find('div', class_='prc')
            if not price_tag:
                price_tag = product.find('span', class_='price')
            price = price_tag.text.strip() if price_tag else "N/A"
            
            
            # Rating
            rating_tag = product.find('div', class_='stars')
            rating = rating_tag.get('data-rating', 'N/A') if rating_tag else "N/A"
            
            
            # Image
            img_tag = product.find('img', class_='img')
            image_url = img_tag.get('data-src') or img_tag.get('src') if img_tag else "N/A"
            
            # Product link
            link_tag = product.find('a', class_='core')
            if not link_tag:
                link_tag = product.find('a', href=True)
            product_link = "https://www.jumia.ci" + link_tag['href'] if link_tag and link_tag.get('href') else "N/A"
            
            # Store extracted data
            product_data.append({
                'name': name,
                'price': price,
                'image_url': image_url,
                'product_link': product_link
            })
            
            if len(product_data) >= max_items:
                break
        
        current_page += 1
        time.sleep(2)  # Respectful delay
        
    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error fetching page {current_page}: {e}")
        break

# Print results
print(f"\n‚úÖ Scraped {len(product_data)} products:\n")
for i, product in enumerate(product_data[:10], 1):
    print(f"{i}. {product['name']}")
    print(f"   Price      : {product['price']}")
    print(f"   Image URL  : {product['image_url']}")
    print(f"   Link       : {product['product_link']}\n")

if len(product_data) > 10:
    print(f"... and {len(product_data) - 10} more products\n")

# Save to CSV
csv_filename = 'jumia_data.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['name', 'price', 'image_url', 'product_link']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    writer.writerows(product_data)

print(f"üíæ Data saved to {csv_filename}")
```

```python
import pandas as pd
from sqlalchemy import create_engine
import mysql.connector

# Method 1: Using SQLAlchemy with correct credentials
def create_mysql_connection():
    try:
        # Replace with your actual MySQL credentials
        username = "root"  # or your MySQL username
        password = "iugb"  # your MySQL password
        host = "localhost"  # or your host
        database = "jumia_products"  # your database name
        
        # Create connection string
        connection_string = f"mysql+pymysql://{username}:{password}@{host}/{database}"
        
        # Create engine
        engine = create_engine(connection_string)
        
        # Test connection
        with engine.connect() as conn:
            print("MySQL connection successful!")
            return engine
            
    except Exception as e:
        print(f"MySQL connection failed: {e}")
        return None

# Create table first (if it doesn't exist)
def create_table(engine):
    create_table_query = """
    CREATE TABLE IF NOT EXISTS products (
        id INT PRIMARY KEY AUTO_INCREMENT,
        name VARCHAR(500) NOT NULL,
        price VARCHAR(100),
        image_url VARCHAR(500),
        product_link VARCHAR(500),
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
    """
    try:
        with engine.connect() as conn:
            conn.execute(create_table_query)
        print("Table created successfully!")
    except Exception as e:
        print(f"Error creating table: {e}")

# Insert data
def insert_data_to_mysql(data):
    engine = create_mysql_connection()
    if engine:
        create_table(engine)
        try:
            data.to_sql('products', con=engine, if_exists='append', index=False)
            print("Data inserted successfully!")
        except Exception as e:
            print(f"Error inserting data: {e}")
    else:
        print("Could not establish database connection")
```
